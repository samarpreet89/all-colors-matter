```python
files = os.listdir(pref)
files_dic = {}
for f in files:
    p = '_'.join(f.split('_')[:-2])
    files_dic[p] = files_dic.get(p, []) + [f]
```

```python
files_small = os.listdir(pref_small)
files_dic_small = {}
for f in files_small:
    p = '_'.join(f.split('_')[:-2])
    files_dic_small[p] = files_dic_small.get(p, []) + [f]
```

```python
len(files), len(files_small)
```

```python
min_cat = None
min_len = 3e5

for p in files_dic_small.keys():
    if p in files_dic.keys():
        l = len(set(files_dic_small[p]).intersection(set(files_dic[p])))
        if l > 0 and l < len(files_dic_small[p]):
            if len(files_dic_small[p]) < min_len:
                min_len = l
                min_cat = p
```
